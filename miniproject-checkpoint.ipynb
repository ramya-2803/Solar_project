```python
# ============================================================
# TRAIN HIGH-ACCURACY XGBOOST MODEL FOR IRRADIANCE PREDICTION
# ============================================================

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, r2_score
from xgboost import XGBRegressor
import joblib

# Load dataset
df = pd.read_excel("FEATURE_ENGINEERED_DATA.xlsx")

# Remove DATE
df = df.drop(columns=["DATE"])

target = "ALLSKY_SFC_SW_DWN"

feature_columns = [
    "MONTH",
    "DAY_OF_YEAR",
    "T2M",
    "T2M_MIN",
    "T2M_MAX",
    "TEMP_AVG",
    "SUN_MEAN_7D",
    "TEMP_MEAN_7D",
    "LAG_1",
    "RH2M",
    "WS2M"
]

X = df[feature_columns]
y = df[target]

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, shuffle=False
)

# =============================
# XGBOOST MODEL (High Accuracy)
# =============================
model = XGBRegressor(
    n_estimators=600,
    learning_rate=0.03,
    max_depth=8,
    subsample=0.9,
    colsample_bytree=0.8,
    reg_alpha=1,
    reg_lambda=1,
    random_state=42,
    tree_method="hist"
)

print("Training XGBoost Model...")
model.fit(X_train, y_train)
print("Training completed!")

# Predictions
y_pred = model.predict(X_test)

# Metrics
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("\n========== MODEL ACCURACY ==========")
print(f"MAE: {mae:.4f}")
print(f"R² Score: {r2:.4f}   (Accuracy ≈ {r2*100:.2f}%)")
print("===================================\n")

# Save model
joblib.dump(model, "solar_model_full.pkl")
print("Model saved as solar_model_full.pkl")
```
